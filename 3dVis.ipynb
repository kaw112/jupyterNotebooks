Visualizing functions of more than 1 dimension. In 2d, if we say y=5, we are implying that for every value of x, y=5.  If we were to write a set of points which belong to the line, every y value would be equal to 5 regardless of the x value (ie {(1,5), (-325,5), (4.5673,5)} are all examples of points that lie on this line).

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10, num = 2)
y = np.array([5,5])

plt.plot(x,y)
plt.show()
```

With a third coordinate--call it z--if we say y=5, we now mean that for every pair of values of x and z , y would be equal to 5.  The set of points corresponding to y=5, then becomes a plane, and we can again pick any point (x,5,z) (for any value of x and z) and know that it belongs to the plane. 



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.clf()
x = np.linspace(0,10, num = 5)
z = np.linspace(0,10, num = 5)
Y = np.ones((5,5))*5.0

X, Z = np.meshgrid(x,z) 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X,Y,Z)
plt.show()
```

This might seem unneccesarily tedious to discuss simple objects in this manner, but thinking about multivariable functions as corresponding to sets of points is a powerful way to help us visualize functions and hence relationships between real world quanities.  Consider the  set of points given by x=5 and y=5 in three dimensions.  The set of points corresponding to these conditions must live in the plane y=5, since every point where y=5 is specified by the plane.  Now, however, we only let x = 5 instead of any value.  This further restricts of the set of points to a line within the plane.   

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.clf()
x = np.linspace(0,10, num = 5)
z = np.linspace(0,10, num = 5)
Y = np.ones((5,5))*5.0

X, Z = np.meshgrid(x,z) 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X,Y,Z)

x1 = np.array([5,5])
y1 = np.array([5,5])
z1 = np.array([0,10])

plt.plot(x1,y1,z1)
plt.show()
```

What do the functions y = 5 and z = x look like?

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact

def Zfunc(x):
    return x

def makePlot(phi, theta):
    plt.clf()
    x = np.linspace(0,10, num = 5)
    z = np.linspace(0,10, num = 5)
    Y = np.ones((5,5))*5.0

    X, Z = np.meshgrid(x,z) 
    fig = plt.figure(figsize=(16, 16), dpi= 80, facecolor='w', edgecolor='k')
    ax = fig.add_subplot(111, projection='3d')

    ax.plot_wireframe(X,Y,Z)

    y2 = np.linspace(0,10, num = 5)

    X1,Y1 = np.meshgrid(x, y2)

    Z1 = Zfunc(X1)

    ax.plot_wireframe(X1,Y1,Z1)

    x1 = np.array([0,10])
    y1 = np.array([5,5])
    z1 = np.array([0,10])

    plt.plot(x1,y1,z1)
    ax.view_init(elev= phi, azim = theta)
    plt.show()
interact(makePlot, phi= (0, 90,0.05), theta=(0,90, 0.5))
```

The fact that x,y, or z = constant represents a plane of unchanging x, y or z value (respectively) can be a useful tool for graphing functions of two variables by hand.  Consider f(x,y) = sin(x)sin(y) shown below.  

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact

def f(x,y):
    return np.sin(x)*np.sin(y)

def plot(phi, theta):
    plt.clf()
    fig = plt.figure(figsize=(16, 16), dpi= 80, facecolor='w', edgecolor='k')
    x = np.linspace(0,2*np.pi, num =30)
    y = np.linspace(0,2*np.pi, num = 30)

    X,Y = np.meshgrid(x,y)

    ax = fig.add_subplot(111, projection='3d')


    ax.plot_wireframe(X,Y,f(X,Y), color='blue')
    ax.view_init(elev= phi, azim = theta)
    plt.show()

interact(plot, phi= (0, 90,0.5), theta=(0,90, 0.5))
```

It might seem like a difficult task to draw this function just by seeing the equation (as one would do for a function of 1 variable), but we can use the use the intersections of planes with the surface to draw curves to "fill in" what the function should look like.  For example consider y = 2.  We know that this is a plane extending forever in the x and z directions but only contains points where y=2.  Hence, what does the set of points resulting from the intersection of y = 2 and f(x,y) = sin(x)sin(y) look like (here we consider that we plot f on the z axis). The curve representing the intersection of the two surfaces is outlined in red.    

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact

def f(x,y):
    return np.sin(x)*np.sin(y)

def g(x):
    return np.sin(x)*np.sin(2.0)

def plot(phi, theta):
    plt.clf()
    fig = plt.figure(figsize=(12, 24), dpi= 80, facecolor='w', edgecolor='k')
    x = np.linspace(0,2*np.pi, num =30)
    y = np.linspace(0,2*np.pi, num = 30)

    X,Y = np.meshgrid(x,y)

    ax = fig.add_subplot(211, projection='3d')
    ax1 = fig.add_subplot(212)
    
    


    ax.plot_wireframe(X,Y,f(X,Y), color='blue')
    z = np.linspace(-2,2, num = 20)
    X1,Z1 = np.meshgrid(x,z)
    Y1 =np.ones(np.shape(X1))*2.0
    ax.plot_wireframe(X1,Y1, Z1, color='red')
    ax.plot(x,np.ones(len(x))*2.0, g(x), linewidth = 3, color='red')
    ax1.plot(x,g(x),linewidth = 3, color='red')
    ax.view_init(elev= phi, azim = theta)
    plt.show()

interact(plot, phi= (0, 90,0.5), theta=(0,90, 0.5))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact

def f(x,y):
    return np.sin(x)*np.sin(y)

def plot(phi, theta):
    plt.clf()
    fig = plt.figure(figsize=(16, 16), dpi= 80, facecolor='w', edgecolor='k')
    x = np.linspace(0,2*np.pi, num =30)
    y = np.linspace(0,2*np.pi, num = 30)


    ax = fig.add_subplot(111, projection='3d')
    for i in range(len(y)):
        ax.plot(x,np.ones(len(x))*y[i], f(x, y[i]), linewidth = 3)
        
    ax.view_init(elev= phi, azim = theta)
    plt.show()

interact(plot, phi= (0, 90,0.5), theta=(0,90, 0.5))
```

What does the function $f(x,y) = y\cdot sin(x)$ look like?

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact

def f(x,y):
    return np.sin(x)*y

def plot(phi, theta):
    plt.clf()
    fig = plt.figure(figsize=(16, 16), dpi= 80, facecolor='w', edgecolor='k')
    x = np.linspace(-np.pi, np.pi, num =30)
    y = np.linspace(-np.pi, np.pi, num = 30)

    X,Y = np.meshgrid(x,y)

    ax = fig.add_subplot(111, projection='3d')
    ax.plot_wireframe(X,Y,f(X,Y))
        
    ax.view_init(elev= phi, azim = theta)
    ax.set_xlim(-np.pi,np.pi)
    ax.set_ylim(-np.pi,np.pi)
    plt.show()

interact(plot, phi= (0, 90,0.5), theta=(0,90, 0.5))
```

If we hold constant values of y and an make cuts through the surface with planes of constant y, we should see that the intersecting curves are sine functions with changing amplitude.  For example $f(x,y=2) = 2 sin(x)$ and $f(x,y=-5) = -5 sin(x)$.  Below we show an animation where we move the plane through different values of y and show the resulting curve.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact
import numpy as np
from matplotlib import animation, rc
from IPython.display import HTML, Image


def f(x,y):
    return np.sin(x)*y


#number of frames to use
frames = 20

#make arrays of data points to evaluate 
x =np.linspace(-np.pi,np.pi,20)
y = np.linspace(-np.pi,np.pi,20)
X,Y = np.meshgrid(x,y)
F = f(X,Y)

z = np.linspace(-np.pi, np.pi, 20)
Z = np.meshgrid(x,z)[1]



def makeAnimation(phi,theta):
    fig = plt.figure(figsize=(10, 16), dpi= 80, facecolor='w', edgecolor='k')
    ax = fig.add_subplot(211,projection='3d')
    ax1 = fig.add_subplot(212)
    ax.view_init(elev= phi, azim = theta)
    
    def init():
        ax.set_xlim(-np.pi,np.pi)
        ax.set_ylim(-np.pi,np.pi)
        ax.set_zlim(-np.pi, np.pi)
        ax1.set_xlim(-np.pi,np.pi)
        ax1.set_ylim(-np.pi,np.pi)
        
        #make wireframe surface
        x =np.linspace(-np.pi, np.pi,20)
        y = np.linspace(-np.pi,np.pi,20)
        X,Y = np.meshgrid(x,y)
    
        S = ax.plot_wireframe(X,Y,F)
        T = ax.plot_wireframe([],[],[])
        U = ax.plot([],[],[])
        V = ax1.plot([],[])
        return S, 
    
    i = 0
    def animate(i):
        ax.clear()
        ax1.clear()

        ax.set_xlim(-np.pi,np.pi)
        ax.set_ylim(-np.pi,np.pi)
        ax.set_zlim(-np.pi, np.pi)
        ax1.set_xlim(-np.pi,np.pi)
        ax1.set_ylim(-np.pi,np.pi)
        
        yVal = -np.pi + (i* 2.0)/float(frames) * np.pi
        Yplane = np.ones((20,20))* yVal
        yCurve = np.ones(20)* yVal
        fCurve = f(x,yCurve)
        
        S = ax.plot_wireframe(X,Y,F)
        T = ax.plot_wireframe(X,Yplane, Z, color = 'red')
        U = ax.plot(x, yCurve,fCurve , color = 'red', linewidth = 3)
        V = ax1.plot(x, fCurve, color = 'red')
        

        return S, 
    rc('animation', html='html5')
    anim = animation.FuncAnimation(fig, animate,init_func=init, frames=frames, interval=180, blit=True)

    return anim

    
anim = interact(makeAnimation, phi= (0, 90,0.05), theta=(0,180, 0.1)) 
anim
```

Thus far we have fixed values of x and y and graphed the resulting curves given by the intersection of the curves and surfaces.  Another way to represent a multivariable function is a contour plot.   Here we can fix values of z, asuming that values of the function are graphed on the z axis, and draw the resulting curves. This is essentially the same thing we have done above, but often times the contour plot is only drawn in the xy plane.   

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact
import numpy as np
from matplotlib import animation, rc
from IPython.display import HTML, Image


def f(x,y):
    return np.exp(-((x)**2+ (y)**2))

def yPlus(x, k):
    return np.sqrt(-np.log(k)-x**2)
    
def yMinus(x, k):
    return -np.sqrt(-np.log(k)-x**2)
    
#number of frames to use
frames = 30

#make arrays of data points to evaluate 
x =np.linspace(-3,3,20)
y = np.linspace(-3,3,20)
X,Y = np.meshgrid(x,y)
F = f(X,Y)




def makeAnimation(phi,theta):
    fig = plt.figure(figsize=(10, 16), dpi= 80, facecolor='w', edgecolor='k')
    ax = fig.add_subplot(211,projection='3d')
    ax1 = fig.add_subplot(212)
    ax.view_init(elev= phi, azim = theta)
    
    def init():
        ax.set_xlim(-3,3)
        ax.set_ylim(-3,3)
        ax.set_zlim(0,1)
        ax1.set_xlim(-3,3)
        ax1.set_ylim(-3,3)
    
        S = ax.plot_wireframe(X,Y,F)
        T = ax.plot_wireframe([],[],[])
        U = ax.plot([],[],[])
        V = ax.plot([],[], [])
        W = ax1.plot([],[])
        R = ax1.plot([],[])
        return S, 
    
    i = 0
    def animate(i):
        ax.clear()
        #ax1.clear()

        ax.set_xlim(-3,3)
        ax.set_ylim(-3,3)
        ax.set_zlim(0,1)
        ax1.set_xlim(-3,3)
        ax1.set_ylim(-3,3)
        
        zVal = 1.0 - (i)/float(frames) 
        Zplane = np.ones((20,20))* zVal
        zCurve = np.ones(35)* zVal
        xCurve = np.linspace(-np.sqrt(-np.log(zVal))+0.0001,np.sqrt(-np.log(zVal))-0.0001, 35)
        yCurvePlus = yPlus(xCurve, zVal)
        yCurveMinus = yMinus(xCurve, zVal)
        
        S = ax.plot_wireframe(X,Y,F)
        T = ax.plot_wireframe(X, Y, Zplane, color = 'red')
        U = ax.plot(xCurve, yCurvePlus,zCurve , color = 'red', linewidth = 3)
        V = ax.plot(xCurve, yCurveMinus, zCurve, color = 'red', linewidth = 3)
        W = ax1.plot(xCurve, yCurvePlus, color = 'red')
        R = ax1.plot(xCurve, yCurveMinus, color = 'red')
        

        return S, 
    rc('animation', html='html5')
    anim = animation.FuncAnimation(fig, animate,init_func=init, frames=frames, interval=180, blit=True)

    return anim

    
anim = interact(makeAnimation, phi= (0, 90,0.05), theta=(0,180, 0.1)) 
anim
```
