```python
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    return np.sin(4*np.pi*t)

def fft(data):
    fft= np.fft.fft(data,6000)
    return fft


plt.clf()
fig, ((ax1,ax2),(ax3,ax4),(fa1,fa2), (fa3,fa4)) = plt.subplots(4,2)
fig.set_size_inches(14, 30)
t1 = np.linspace(0,16,800)
t2 = np.linspace(0,8, 400)
t3 = np.linspace(0,4, 200)
t4 = np.linspace(0, 2, 100)
plt.xlim(0,16)

ax1.plot(t1, f(t1))
ax2.plot(t2, f(t2))
ax3.plot(t3, f(t3))
ax4.plot(t4, f(t4))
ax1.set_xlim(0,16)
ax2.set_xlim(0,16)
ax3.set_xlim(0,16)
ax4.set_xlim(0,16)


freqArray = np.fft.fftfreq(6000, 1.0/(2*np.pi*50.0))
ft1 = fft(f(t1))
ft2 = fft(f(t2))
ft3 = fft(f(t3))
ft4 = fft(f(t4))

#f1.set_xlim(10,15)
#f2.set_xlim(10,15)
#f3.set_xlim(10,15)
#f4.set_xlim(10,15)
#f1.set_ylim(-300,300)
#f2.set_ylim(-300,300)
#f3.set_ylim(-300,300)
#f4.set_ylim(-300,300)
#fi1.set_xlim(10,15)
#fi2.set_xlim(10,15)
#fi3.set_xlim(10,15)
#fi4.set_xlim(10,15)
#fi1.set_ylim(-400,100)
#fi2.set_ylim(-400,100)
#fi3.set_ylim(-400,100)
#fi4.set_ylim(-400,100)
fa1.set_xlim(10,15)
fa2.set_xlim(10,15)
fa3.set_xlim(10,15)
fa4.set_xlim(10,15)
fa1.set_ylim(0,160000)
fa2.set_ylim(0,40000)
fa3.set_ylim(0,20000)
fa4.set_ylim(0,10000)

#f1.plot(freqArray, ft1.real)
#f2.plot(freqArray, ft2.real)
#f3.plot(freqArray, ft3.real)
#f4.plot(freqArray, ft4.real)
#fi1.plot(freqArray, ft1.imag)
#fi2.plot(freqArray, ft2.imag)
#fi3.plot(freqArray, ft3.imag)
#fi4.plot(freqArray, ft4.imag)
fa1.plot(freqArray, ft1.real**2+ft1.imag**2)
fa2.plot(freqArray, ft2.real**2+ft2.imag**2)
fa3.plot(freqArray, ft3.real**2+ft3.imag**2)
fa4.plot(freqArray, ft4.real**2+ft4.imag**2)

ax1.set_ylabel('f1(t)', fontsize = 13)
ax2.set_ylabel('f2(t)', fontsize = 13)
ax3.set_ylabel('f3(t)',fontsize = 13)
ax4.set_ylabel('f4(t)',fontsize = 13)
ax3.set_xlabel('t',fontsize = 13)
ax4.set_xlabel('t', fontsize = 13)
#f1.set_ylabel(r'Re(F1( w))', fontsize = 13)
#f2.set_ylabel(r'Re(F2( w))', fontsize = 13)
#f3.set_ylabel(r'Re(F3(w))',fontsize = 13)
#f4.set_ylabel(r'Re(F4(w))',fontsize = 13)
fa3.set_xlabel(r'w',fontsize = 13)
fa4.set_xlabel(r'w', fontsize = 13)
#fi1.set_ylabel(r'Im(F1(w))', fontsize = 13)
#fi2.set_ylabel(r'Im(F2(w))',fontsize = 13)
#fi3.set_ylabel(r'Im(F3(w))',fontsize = 13)
#fi4.set_ylabel(r'Im(F4(w))',fontsize = 13)
fa1.set_ylabel(r'$|F1(w)|^2$', fontsize = 13)
fa2.set_ylabel(r'$|F2(w)|^2$',fontsize = 13)
fa3.set_ylabel(r'$|F3(w)|^2$',fontsize = 13)
fa4.set_ylabel(r'$|F4(w)|^2$',fontsize = 13)
plt.show()
```
